# Histogram

A histogram represents the distribution of a variable (over a specific range) by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin.

```python

flowers_df.info()           
```

```python
plt.title("Distribution of petal_length")
plt.hist(flowers_df.petal_length)
plt.legend(["petal_length"])
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

# Multiple Histogram

simialr to line charts as we have drawn two or more line chart in single plot to analyze the relation

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
setosa_df
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.legend(["setosa","versicolour"])
```

```python
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
```
